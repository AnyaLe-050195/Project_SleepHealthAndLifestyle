## Purpose of the project

The "Sleep Health and Lifestyle" project aims to shed light on the critical role of sleep in overall health and well-being. By exploring and visualizing data related to sleep habits and lifestyle factors, I want to raise awareness about the importance of prioritizing sleep for maintaining optimal health. 

Through data exploration, my goal is to identify patterns and trends in sleep behaviors and lifestyle choices. By doing so, I aim to offer valuable insights that individuals can use to make informed decisions about their health and lifestyle.

## Data features

1. **Person ID**: Unique identifier assigned to each individual in the dataset.
2. **Gender**: Categorical variable indicating the gender of the individual (e.g., Male, Female, Other).
3. **Age**: Age of the individual in years.
4. **Occupation**: The type of work or profession the individual is engaged in.
5. **Sleep Duration**: The average number of hours of sleep per day for the individual.
6. **Quality of Sleep**: A subjective rating on a scale of 1 to 10, indicating the perceived quality of sleep.
7. **Physical Activity Level**: The average number of minutes the person engages in physical activity daily.
8. **Stress Level**: A subjective rating on a scale of 1 to 10, indicating the perceived level of stress.
9. **BMI Category**: Body Mass Index (BMI) category indicating the individual's weight status (e.g., Underweight, Normal weight, Overweight, Obese).
10. **Blood Pressure**: The individual's blood pressure reading, typically indicated as systolic pressure over diastolic pressure (e.g., 120/80 mmHg).
11. **Heart Rate**: The individual's heart rate measured in beats per minute.
12. **Daily Steps**: The average number of steps taken by the individual per day.
13. **Sleep Disorder**: Categorical variable indicating whether the individual has a sleep disorder, with possible values of `None`, `Insomnia`, or `Sleep Apnea`.

## Data exploratory

```python
# Columns name
col_name = sleep_data.columns
print(col_name)
```

```python
# Percentage of persons have sleep disorder or not
pct_sleep_disorder = pd.DataFrame(sleep_data["Sleep Disorder"].value_counts(sort=True))
pct_sleep_disorder.columns = ["Count"]
pct_sleep_disorder["Percent"] = round(sleep_data["Sleep Disorder"].value_counts(normalize=True)*100,2)
print(pct_sleep_disorder)
print(f"\nObservation: \nThe proportion of {pct_sleep_disorder[pct_sleep_disorder==pct_sleep_disorder.max()].index[0]} is more")
```

```python
# Percentage of people with sleep problems or not classified by gender
pct_sleep_gender = pd.DataFrame(sleep_data.groupby("Sleep Disorder")["Gender"].value_counts().unstack(level=1))
#print(pct_sleep_gender)

## Create bar plot
ax = pct_sleep_gender.plot(kind="bar",
                           stacked=True,
                           title="Sleep Disorders by Gender",
                           xlabel="Sleep Disorder",
                           ylabel="Count",
                           legend=True,
                           figsize=(3.5,7),
                           width=0.5,
                           color={"Female": "#F3CF39",
                                  "Male": "#648E5B"})     

### Sum each row to get total
totals = pct_sleep_gender.sum(axis=1)

### Set offset to bump total label up a bit above the bar
y_offset_1 = 3

### Add total label 
for i, total in enumerate(totals):
      ax.text(i, total + y_offset_1, total, ha="center", weight="bold", size = 10)

### Set offset to put label in the middle of each sub-bar
y_offset_2 = -7
### Add label for each sub-bar
for bar in ax.patches:
      ax.text(
            # Put the text in the middle of each bar. get_x returns the start
            bar.get_x() + bar.get_width() / 2,
            # Vertically, add the height of bar to the start of bar, along with the offset.
            bar.get_height() + bar.get_y() + y_offset_2,
            # Actual value to show
            bar.get_height(),
            # Style
            ha="center",
            weight="bold",
            color='w',
            size=7
      ) 

plt.show()

## Compare and print Observation
print("\nObservation: ")
comparison = pct_sleep_gender["Female"] > pct_sleep_gender["Male"]
for ind, val in comparison.items():
      if val:
            print(f"For {ind}, Female is greater than Male")
      else:
            print(f"For {ind}, Female is less than or equal to Male")
```

```python
# Percentage of people with sleep problems or not classified by occupation
pct_sleep_job = pd.DataFrame(sleep_data.groupby("Sleep Disorder")["Occupation"].value_counts())
pct_sleep_job.columns = ["Count"]
pct_sleep_job = pct_sleep_job.reset_index()
#print(pct_sleep_job)

## Create tree map
fig = px.treemap(data_frame=pct_sleep_job,
                 values="Count",
                 width=700,
                 height=700,
                 color="Sleep Disorder",
                 color_discrete_sequence=["#FFC3A1", "#F0997D","#A75D5D", "#FFC3A1"],
                 path=[px.Constant("Jobs"), "Sleep Disorder", "Occupation"],
                 title = "Sleep Disorders by Occupation")
fig.show()

## Compare and print Observation
max_job = pct_sleep_job.groupby("Sleep Disorder")["Count"].max()
max_job_name = []
for val in max_job:
      max_job_name.append(pct_sleep_job[pct_sleep_job["Count"]==val][["Sleep Disorder","Occupation"]])
print("\nObservation: ")
for e in max_job_name:
      print(f"For {e.iloc[0][0]}, {e.iloc[0][1]} more than others")
```

```python
# Correlation between sleep duration, sleep quality and sleep disorder
## Correlation between sleep duration, sleep quality
corr_dur_qual = sleep_data[["Sleep Duration", "Quality of Sleep"]].corr()
print(corr_dur_qual)

## Print Observation
print("\nObservation:\nBased on the values in the variable corr_dur_qual:")
if corr_dur_qual.loc["Sleep Duration", "Quality of Sleep"] > 0:
    print("It means that the longer the sleep duration, the better the quality of sleep and vice versa")
elif 0 > corr_dur_qual.loc["Sleep Duration", "Quality of Sleep"] <= 0:
    print("It means that as the sleep duration or the quality of sleep increases, the other variable may decrease and vice versa")  

## The effect of sleep duration, sleep quality on sleep disorder
### Check outliers data in Sleep Duration column and print result
iqr_dur = iqr(sleep_data["Sleep Duration"])
lower_threshold_dur = np.quantile(sleep_data["Sleep Duration"], 0.25) - 1.5 * iqr_dur
upper_threshold_dur = np.quantile(sleep_data["Sleep Duration"], 0.75) + 1.5 * iqr_dur
outliers_dur = sleep_data[(sleep_data["Sleep Duration"] < lower_threshold_dur) | (sleep_data["Sleep Duration"] > upper_threshold_dur)]
print("\nBased on the IQR method, the outliers in Sleep Duration are:")
if outliers_dur.shape[0] == 0:
    print("No outliers in Sleep Duration")
else:
    print(outliers_dur)

### Create pivot table to compare number of people with sleep disorder by quality of sleep
pivot_person_qual_disorder = sleep_data.pivot_table(index = "Quality of Sleep",
                                                    columns = "Sleep Disorder",
                                                    values = "Person ID",
                                                    aggfunc = "count",
                                                    fill_value = 0)
#print(pivot_person_qual_disorder)

### Create bar plot to compare number of people with sleep disorder by quality of sleep
pivot_person_qual_disorder.plot(kind="bar",
                                title="Quality of Sleep by Sleep Disorder",
                                xlabel="Quality of Sleep",
                                ylabel="Count",
                                figsize=(10,5),
                                legend=True,
                                color={"Insomnia": "#FFC3A1",
                                       "None": "#F0997D",
                                       "Sleep Apnea": "#A75D5D"})
plt.show()

### Create pivot table to compare sleep duration by quality of sleep and sleep disorder
pivot_dur_qual_disorder = sleep_data.pivot_table(index = "Quality of Sleep",
                                                 columns = "Sleep Disorder",
                                                 values = "Sleep Duration",
                                                 fill_value = 0)
#print(pivot_dur_qual_disorder)

### Create bar plot to compare sleep duration by quality of sleep and sleep disorder
pivot_dur_qual_disorder.plot(kind = "bar", 
                             title = "Sleep Duration by Quality of Sleep and Sleep Disorder",
                             xlabel = "Quality of Sleep",
                             ylabel = "Sleep Duration",
                             figsize = (10,5),
                             legend = True,
                             color = {"Insomnia": "#FFC3A1",
                                      "None": "#F0997D",
                                      "Sleep Apnea": "#A75D5D"})
plt.show()
```

<font size="2.5">**Observation based on the Quality of Sleep by Sleep Disorder bar chart:**</font>

<font size="2.5">1. **Insomnia:**</font>

   - <font size="2.5">People with insomnia tend to have lower sleep quality than those without the condition, peaking at levels 6 and 7.</font>

<font size="2.5">2. **None (No sleep disorder):**</font>

   - <font size="2.5">Individuals without any sleep disorders typically have the highest sleep quality, predominantly at level 8.</font>

   - <font size="2.5">This indicates that those without sleep disorders are more likely to experience deep and restful sleep, leading to better sleep quality.</font>

<font size="2.5">3. **Sleep Apnea:**</font>

   - <font size="2.5">The number of people with sleep apnea shows no significant variation across different levels of sleep quality.</font>

   - <font size="2.5">This suggests that sleep apnea may not significantly impact sleep quality, as individuals with this disorder are evenly distributed across different levels of sleep quality.</font>


<font size="2.5">**Observation based on the Sleep Duration by Quality of Sleep and Sleep Disorder:**</font>

<font size="2.5">1. **Insomnia:**</font>

   - <font size="2.5">The average sleep duration increases with lower sleep quality compared to the other group.</font>

<font size="2.5">2. **None (No sleep disorder):**</font>

   - <font size="2.5">The average sleep duration remains consistently high (ranging from 6.12 to 8.37) across all levels of sleep quality.</font>
   
<font size="2.5">3. **Sleep Apnea:**</font>

   - <font size="2.5">The average sleep duration fluctuates inconsistently.</font>

```python
# The effect of age, gender on sleep disorder
## Calculate the age group for 'Sleep Disorder' and 'Gender' columns
age_group = sleep_data.groupby(["Sleep Disorder", "Gender"], group_keys=False)["Age"].apply(lambda x: pd.cut(x, bins=[0, 18, 30, 50, float('inf')], labels=["<18", "18-30", "30-50", ">50"]))

## Percentage of sleep disorder classified by age group
age_group_count = age_group.value_counts()

### Create pie chart
plt.pie(age_group_count, 
        labels=age_group_count.index.values, 
        autopct='%1.1f%%',
        startangle=150, 
        pctdistance = 0.8,
        explode=(0.1, 0.0, 0.2, 0.4),
        shadow=True,
        colors=["#F0A538", "#D99B76", "#C2CC8F", "brown"])
plt.legend(title="Age Group", loc="upper right", bbox_to_anchor=(0.75, 0, 0.5, 1))
plt.title("Age Group by Sleep Disorder")
plt.show()

## Find the age group most affected for each sleep disorder and gender
most_affected_age_group = age_group.groupby([sleep_data['Sleep Disorder'], sleep_data['Gender']]).apply(lambda x: x.value_counts().idxmax())

## Compare and print Observation
print("\nObservation: ")
print(f"The age group most affected by sleep disorders is {age_group_count.idxmax()}")
print(f"Age Group Most Affected for Each Sleep Disorder and Gender:\n{most_affected_age_group.to_string(header=False)}")

## Create ECDF chart showing the probability of sleep disorders distributed by age and gender
fig_ecdf_sex_age = px.ecdf(sleep_data, 
                           x="Age", 
                           facet_col="Sleep Disorder",
                           title="The effect of age, gender on sleep disorder",
                           color="Gender",
                           color_discrete_sequence=["#648E5B", "#F3CF39"],
                           markers=True)
fig_ecdf_sex_age.update_layout(plot_bgcolor="seashell")
fig_ecdf_sex_age.update_xaxes(showgrid=True, gridwidth=2)
fig_ecdf_sex_age.show()
```

<font size="2.5">**Observation based on the The effect of age, gender on sleep disorder:**</font>

<font size="2.5">1. **Insomnia:**</font>

   - <font size="2.5">Before the age of 44, the likelihood of insomnia in males and females is almost the same.</font>
   - <font size="2.5">After reaching the age of 44, the likelihood of insomnia increases in both males and females.</font>

<font size="2.5">2. **None (No sleep disorder):**</font>

   - <font size="2.5">Under the age of 45, the likelihood of males not experiencing sleep problems is 2 to 3 times greater than that of females.</font>
   
<font size="2.5">3. **Sleep Apnea:**</font>

   - <font size="2.5">The likelihood of men suffering from sleep apnea surpasses that of women, particularly noticeable from the age of 38 or older, where men are 5 to 6 times more likely to develop the condition compared to women.</font>
   - <font size="2.5">At an equivalent probability level, men tend to develop the disease at a younger age than women. For instance, with a probability of 60, men typically develop the condition at 38 years old, while women develop it at around 57 years old.</font>
   - <font size="2.5">Starting from the age of 50 onwards, the probability of women acquiring the disease accelerates more rapidly than before. Meanwhile, in men, the likelihood of developing the disease steadily increases each year.</font>

```python
# The effect of Physical Activity Level on sleep disorder
## Check outliers data in Physical Activity Level column and print result
iqr_physic = iqr(sleep_data["Physical Activity Level"])
lower_threshold_physic = np.quantile(sleep_data["Physical Activity Level"], 0.25) - 1.5 * iqr_physic
upper_threshold_physic = np.quantile(sleep_data["Physical Activity Level"], 0.75) + 1.5 * iqr_physic
outliers_physic = sleep_data[(sleep_data["Physical Activity Level"] < lower_threshold_physic) | (sleep_data["Physical Activity Level"] > upper_threshold_physic)]
print("\nBased on the IQR method, the outliers in Physical Activity Level are:")
if outliers_physic.shape[0] == 0:
    print("No outliers in Physical Activity Level")
else:
    print(outliers_physic)

## Correlation between Physical Activity Level, sleep quality
corr_physic_qual = sleep_data[["Physical Activity Level", "Quality of Sleep"]].corr()
print(corr_physic_qual)

## Print Observation
print("\nObservation:\nBased on the values in the variable corr_physic_qual:")
if corr_physic_qual.loc["Physical Activity Level", "Quality of Sleep"] > 0:
    print("It means that the higher level of physical activity, the better quality of sleep and vice versa")
elif 0 > corr_physic_qual.loc["Physical Activity Level", "Quality of Sleep"] <= 0:
    print("It means that as the level of physical activity or the quality of sleep increases, the other variable may decrease and vice versa") 

## Create pivot table to compare Physical Activity Level by quality of sleep and sleep disorder
pivot_physic_qual = sleep_data.pivot_table(index = "Quality of Sleep",
                                           columns = "Sleep Disorder",
                                           values = "Physical Activity Level",
                                           fill_value = 0)
print(pivot_physic_qual)

### Create bar plot to compare sleep duration by quality of sleep and sleep disorder
pivot_physic_qual.plot(kind = "bar", 
                       title = "Physical Activity Level by Quality of Sleep and Sleep Disorder",
                       xlabel = "Quality of Sleep",
                       ylabel = "Physical Activity Level",
                       figsize = (10,5),
                       legend = True,
                       color = {"Insomnia": "#FFC3A1",
                                "None": "#F0997D",
                                "Sleep Apnea": "#A75D5D"})
plt.show()
```
